```python
import torch
from collections import OrderedDict

# 1. 加载权重文件（用 CPU 加载更安全）
path = "weights/Expv8_large_HQEVFI.pt"
ckpt = torch.load(path, map_location="cpu")

print(ckpt.keys())
print(ckpt['epoch'])
sd: OrderedDict = ckpt["model_state"]
# print("=== First 50 params in model_state ===")
for i, (name, param) in enumerate(sd.items()):
    # if i >= 50:  # 避免一次性打印太多，看个大概结构
    #     break
    if torch.is_tensor(param):
        print(f"{i:03d}  {name:60s}  {tuple(param.shape)}")
    else:
        print(f"{i:03d}  {name:60s}  (type={type(param)})")
```
